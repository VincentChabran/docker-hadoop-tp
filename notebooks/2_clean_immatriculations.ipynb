# Ajout de la catégorie du véhicule.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.clustering import KMeans
from pyspark.sql.functions import when
```

```python
from pyspark.sql import SparkSession

# Initialisation de la session Spark avec support Hive
spark = SparkSession.builder \
    .appName("Analyse Immatriculations") \
    .config("spark.hadoop.hive.metastore.uris", "thrift://hive-metastore:9083") \
    .enableHiveSupport() \
    .getOrCreate()

# Utilisation de la base Hive et chargement des données d'immatriculations
spark.sql("USE concessionnaire")
immatriculation_df 

# Afficher un aperçu des données
immatriculation_df.show(10)
```

```python

# Transformation des colonnes pour le clustering
features_df = immatriculation_df.select(
    col("Immatriculation"),  # Conserver la clé pour la jointure
    col("Puissance").cast("double"),
    col("Prix").cast("double"),
    col("NbPlaces").cast("double"),
    when(col("Longueur") == "courte", 1)
      .when(col("Longueur") == "moyenne", 2)
      .when(col("Longueur") == "longue", 3)
      .when(col("Longueur") == "très longue", 4).alias("Longueur_numeric")
)

# Afficher un aperçu des données transformées
features_df.show(10)
```

```python
from pyspark.ml.feature import VectorAssembler, StandardScaler

# Création d'un vecteur pour les fonctionnalités
assembler = VectorAssembler(inputCols=["Puissance", "Prix", "NbPlaces", "Longueur_numeric"], outputCol="features")
assembled_df = assembler.transform(features_df)

# Normalisation des données
scaler = StandardScaler(inputCol="features", outputCol="scaled_features")
scaler_model = scaler.fit(assembled_df)
scaled_df = scaler_model.transform(assembled_df)

# Affichage des données assemblées et normalisées
scaled_df.select("Immatriculation", "features", "scaled_features").show(10, truncate=False)
```

```python
from pyspark.ml.clustering import KMeans

# Application de k-means avec 4 clusters
kmeans = KMeans(k=4, seed=42, featuresCol="scaled_features", predictionCol="Cluster")
kmeans_model = kmeans.fit(scaled_df)
clustered_df = kmeans_model.transform(scaled_df)

# Afficher un aperçu des données avec les clusters
clustered_df.select("Immatriculation", "Cluster").show(10)
```

```python
# Analyse des clusters pour comprendre leurs caractéristiques
cluster_analysis = clustered_df.groupBy("Cluster").agg(
    {"Puissance": "avg", "Prix": "avg", "Longueur_numeric": "avg"}
).withColumnRenamed("avg(Puissance)", "Puissance Moyenne") \
 .withColumnRenamed("avg(Prix)", "Prix Moyen") \
 .withColumnRenamed("avg(Longueur_numeric)", "Longueur Moyenne")

# Afficher les caractéristiques des clusters
cluster_analysis.show(truncate=False)
```

```python
# Mapper les clusters aux catégories appropriées en fonction des résultats de l'analyse
clustered_df = clustered_df.withColumn(
    "Catégorie",
    when(col("Cluster") == 0, "Citadine")  # Ajustez ces mappings en fonction des résultats de l'analyse
    .when(col("Cluster") == 1, "Familiale")
    .when(col("Cluster") == 2, "Sportive")
    .when(col("Cluster") == 3, "Luxe")
)

# Vérifiez les nouvelles catégories après ajustement
clustered_df.select("Immatriculation", "Puissance", "Prix", "Longueur_numeric", "Cluster", "Catégorie").show(10)
```

```python
# Joindre les catégories corrigées avec le DataFrame original
immatriculation_with_categories = immatriculation_df.join(
    clustered_df.select("Immatriculation", "Catégorie"), on="Immatriculation", how="left"
)

# Afficher un aperçu des données finales avec les catégories corrigées
immatriculation_with_categories.show(1000, truncate=False)
```
