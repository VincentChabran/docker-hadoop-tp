```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.clustering import KMeans
from pyspark.sql.functions import when
import matplotlib.pyplot as plt
```

# Utilisation de la base de données Hive, et charge les données de la table

```python
spark = SparkSession.builder \
    .appName("Categorie") \
    .config("spark.hadoop.hive.metastore.uris", "thrift://hive-metastore:9083") \
    .enableHiveSupport() \
    .getOrCreate()
```

```python
spark.sql("USE concessionnaire")
client_immat_df = spark.sql("SELECT * FROM client_immatriculation_merged")
```

```python
client_immat_df.show(10, truncate=False)
```

# Définis les catégories.

```python
client_immat_df = client_immat_df.withColumn(
    "categorie",
    when((col("puissance") <= 100) & (col("longueur") == "courte"), "citadine")
    .when((col("puissance") > 100) & (col("puissance") <= 200) & (col("longueur") == "courte") & (col("prix") < 30000), "citadine premium")
    .when((col("puissance") > 100) & (col("puissance") <= 200) & (col("longueur").isin("moyenne", "longue")), "familiale")
    .when((col("puissance") > 200) & (col("prix") > 30000), "sportive")
    .when((col("nbplaces") >= 7), "monospace")
    .when((col("prix") > 50000) & col("marque").isin("BMW", "Mercedes", "Jaguar", "Audi", "Porsche"), "luxe")
    .otherwise("autre")
)
```

# Annalyse du resultat 

```python
client_immat_df.select("marque","modele","categorie","prix","longueur","puissance").show(30, truncate=False)
```

```python
# Étape 1 : Calculer la distribution des catégories
categorie_counts = (
    client_immat_df.groupBy("categorie")
    .count()
    .withColumn("percentage", (col("count") / client_immat_df.count()) * 100)
)

# Afficher les résultats dans la console
categorie_counts.show(truncate=False)
```

```python
# Étape 2 : Convertir en Pandas pour Matplotlib
categorie_counts_pd = categorie_counts.toPandas()

# Étape 3 : Trier les données pour une meilleure visualisation
categorie_counts_pd = categorie_counts_pd.sort_values(by="percentage", ascending=False)

# Étape 4 : Création du graphique
plt.figure(figsize=(8, 5))
plt.bar(
    categorie_counts_pd["categorie"], 
    categorie_counts_pd["percentage"]
)

plt.xlabel("Catégories")
plt.ylabel("Pourcentage (%)")
plt.title("Répartition des Catégories dans le DataFrame")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

plt.show()
```

```python
# Création du graphique camembert
plt.figure(figsize=(6, 6))
plt.pie(
    categorie_counts_pd["percentage"], 
    labels=categorie_counts_pd["categorie"], 
    autopct="%1.1f%%", 
    startangle=140
)

plt.title("Répartition des Catégories dans le DataFrame (en %)")
plt.axis("equal")  # Assure que le graphique est un cercle
plt.tight_layout()

# Afficher le graphique
plt.show()
```

**Observation**
En analysant les données, nous pouvons constater que certaines catégories, comme "citadine premium" et "autre", sont significativement sous-représentées par rapport à d'autres catégories telles que "familiale" ou "sportive".

**Proposition**
Il serait pertinent avec plus de temps de chercher des moyens pour mieux uniformiser la répartition des catégories, notamment en :

- Réévaluant les critères de catégorisation pour redistribuer les véhicules dans des classes plus équilibrées.
- Envisageant une fusion de certaines catégories sous-représentées avec des catégories proches pour simplifier l’analyse et réduire les déséquilibres.
- Augmentant la taille des données pour les catégories sous-représentées, par exemple via des données supplémentaires ou des techniques de génération de données synthétiques.
